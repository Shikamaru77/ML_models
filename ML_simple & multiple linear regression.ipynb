```python
import pandas as pd
import numpy as np
```

```python
read= pd.read_csv('insurance.csv')
```

```python
read
```

```python
read['sex']= read['sex'].astype('category')
read['sex']= read['sex'].cat.codes
read['smoker']= read['smoker'].astype('category')
read['smoker']= read['smoker'].cat.codes
read['region']= read['region'].astype('category')
read['region']= read['region'].cat.codes
read
```

```python
read.isnull().sum()
```

```python
X =read.drop(columns='charges')
X
```

```python
y= read['charges']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
C= lr.intercept_
C
```

```python
m= lr.coef_
m
```

```python
y_pred_train= lr.predict(X_train)
y_pred_train
```

```python
import matplotlib.pyplot as plt
plt.scatter(y_train,y_pred_train)
plt.xlabel("actual")
plt.xlabel("pred")
plt.show()
```

```python
from sklearn.metrics import r2_score
r2_score(y_train,y_pred_train)
```

```python
y_pred_test= lr.predict(X_test)
import matplotlib.pyplot as plt
plt.scatter(y_test,y_pred_test)
plt.xlabel("actual")
plt.xlabel("pred")
plt.show()
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred_test)
```
